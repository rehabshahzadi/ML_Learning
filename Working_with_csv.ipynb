**Get Dataset from Scikit learn**

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes(as_frame=True)
print(diabetes.DESCR)
```

**Transform Dataset into Pandas Dataframe**

```python
import pandas as pd
# Convert the data to a pandas DataFrame
df = pd.DataFrame(
data=diabetes.data,
columns=diabetes.feature_names
)
```

**Checking the Data**

**Store the Dataframe  as a csv file**

```python
df.to_csv('diabetes_dataset.csv', index=False)
```

**Read the csv file as a Dataframe**

```python
df_read = pd.read_csv('diabetes_dataset.csv', nrows=2)
display(df_read.head())
```

```python
df_read = pd.read_csv('diabetes_dataset.csv', skiprows=1)
display(df_read.head())
```

```python
df_read = pd.read_csv('diabetes_dataset.csv', skiprows=[3,2])
display(df_read.head())
```
